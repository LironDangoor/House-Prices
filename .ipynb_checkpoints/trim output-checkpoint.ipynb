```python
#load packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#import data from file
train_data=pd.read_csv ("train.csv")
test_data=pd.read_csv ("test.csv")
train_data.head()
```

```python
sns.catplot(x="Utilities", y="SalePrice", data=train_data, kind="boxen")

#all the house excepy 1 have all utilites- I will drop this columns
```

```python
all_data = pd.concat([train_data, test_data])
```

```python
all_data.drop(columns=['Utilities', 'Id'], inplace=True)
```

```python
all_data.shape
```

```python
total_missing= all_data.isnull().sum()>0
total_missing = total_missing.drop(total_missing[total_missing == 0].index)
print(total_missing)
```

```python
#Decoding Categorial Feature according to their importance
all_data.replace({"FireplaceQu":{np.NaN:0, "Po":1, "Fa":2, "TA":3, "Gd":4, "Ex":5}}, inplace=True)
all_data.replace({"KitchenQual":{"Po":1, "Fa":2, "TA":3, "Gd":4, "Ex":5}}, inplace=True)
all_data.replace({"ExterQual":{"Po":1, "Fa":2, "TA":3, "Gd":4, "Ex":5}}, inplace=True)
all_data.replace({"ExterCond":{"Po":1, "Fa":2, "TA":3, "Gd":4, "Ex":5}}, inplace=True)
all_data.replace({"BsmtQual":{np.NaN:0,"Po":1, "Fa":2, "TA":3, "Gd":4, "Ex":5}}, inplace=True)
all_data.replace({"BsmtCond":{np.NaN:0,"Po":1, "Fa":2, "TA":3, "Gd":4, "Ex":5}}, inplace=True)
all_data.replace({"HeatingQC":{"Po":1, "Fa":2, "TA":3, "Gd":4, "Ex":5}}, inplace=True)
all_data.replace({"FireplaceQu":{np.NaN:0,"Po":1, "Fa":2, "TA":3, "Gd":4, "Ex":5}}, inplace=True)
all_data.replace({"PoolQC":{np.NaN:0,"Po":1, "Fa":2, "TA":3, "Gd":4, "Ex":5}}, inplace=True)
all_data.replace({"GarageQual":{np.NaN:0,"Po":1, "Fa":2, "TA":3, "Gd":4, "Ex":5}}, inplace=True)
all_data.replace({"GarageCond":{np.NaN:0,"Po":1, "Fa":2, "TA":3, "Gd":4, "Ex":5}}, inplace=True)
all_data.replace({"GarageFinish":{np.NaN:0,"Unf":1, "RFn":2, "Fin":3}}, inplace=True)
all_data.replace({"Alley":{np.NaN:0,"Grvl":1, "Pave":2}}, inplace=True)
all_data.replace({"Street":{np.NaN:0,"Grvl":1, "Pave":2}}, inplace=True)
all_data.replace({"BsmtExposure":{np.NaN:0,"No":1, "Mn":2, "Av":3, "Gd":4}}, inplace=True)
all_data.replace({"BsmtFinType1":{np.NaN:0,"Unf":1, "LwQ":2, "Rec":3, "BLQ":4, "ALQ":5, "GLQ":6}}, inplace=True)
all_data.replace({"BsmtFinType2":{np.NaN:0,"Unf":1, "LwQ":2, "Rec":3, "BLQ":4, "ALQ":5, "GLQ":6}}, inplace=True)
all_data.replace({"Fence":{np.NaN:0,"MnWw":1, "GdWo":2, "MnPrv":3, "GdPrv":4}}, inplace=True)

#GarageFinishl decoding
all_data.replace(["Y", "N", "P"], [1,0, 0.5], inplace=True)

#Electrical decoding
all_data.replace(["Mix", "FuseP","FuseF", "FuseA", "SBrkr"], [0,1,2,3,4], inplace=True)

#Functional decoding
all_data.replace(["Sal", "Sev","Maj2", "Maj1", "Mod", "Min2", "Min1", "Typ"], [0,1,2,3,4,5,6,7], inplace=True)

#condition decoding. combine the 2 condition into 1
all_data.replace([ "Artery", "Feedr", "RRNn", "RRAn","PosN", "PosA", "RRNe", "RRAe"], -1, inplace=True)
all_data.replace([ "Norm"], 0, inplace=True)
all_data.replace(["PosN", "PosA"], 1, inplace=True)
all_data["Condition"]= all_data["Condition1"]+ all_data["Condition2"]
all_data.drop(columns=["Condition1", 'Condition2'],inplace=True)
```

```python
# Dealing with missing information- categorial value

#GarageCond- if there is garage- fill the missing value with the most common one
m= all_data['GarageCond'].value_counts().idxmax()
all_data["GarageCond"][(all_data["GarageArea"]>0) & (all_data["GarageCond"].isnull())]=m

#GarageQual- if there is garage- fill the missing value with the most common one
m= all_data['GarageQual'].value_counts().idxmax()
all_data['GarageQual'][(all_data["GarageArea"]>0) & (all_data["GarageQual"].isnull())]=m

#GarageFinishl- if there is garage- fill the missing value with the most common one
m= all_data['GarageFinish'].value_counts().idxmax()
all_data['GarageFinish'][(all_data["GarageArea"]>0) & (all_data["GarageFinish"].isnull())]=m

#Electrical, Functional, KitchenQual, MSZoning, Exterior1st, SaleType
all_data["Electrical"].fillna(all_data['Electrical'].mode()[0], inplace=True)
all_data["Functional"].fillna(all_data['Functional'].mode()[0], inplace=True)
all_data["KitchenQual"].fillna(all_data['KitchenQual'].mode()[0], inplace=True)
all_data["MSZoning"].fillna(all_data['MSZoning'].mode()[0], inplace=True)
all_data["Exterior1st"].fillna(all_data['Exterior1st'].mode()[0], inplace=True)
all_data["SaleType"].fillna(all_data['SaleType'].mode()[0], inplace=True)
```

```python
#fill LotFrontage-null with the most average value
all_data["LotFrontage"].fillna(all_data["LotFrontage"].mean(), inplace=True)
```

```python
#Dealing with missing data in MasVnrType and MasVnrArea
print (all_data["MasVnrType"].unique())
```

```python
# if MasVnrArea>0 replace MasVnrType-null in the most common value
n= all_data[all_data["MasVnrType"]!="None"]["MasVnrType"].value_counts().idxmax()
all_data["MasVnrType"][(all_data["MasVnrArea"] > 0) & (all_data["MasVnrType"]=="None")]=n
all_data["MasVnrType"][(all_data["MasVnrType"].isnull()) & (all_data["MasVnrArea"].isnull()==False) ]=n
# Than, replace all the "NA" in None
all_data["MasVnrType"].fillna('None', inplace=True)
```

```python
# filled the nulls in these columns with 0 since a null in these features means there is None and these are numeric
# BsmtFinSF1, BsmtFinSF2, BsmtUnfSF, BsmtFullBath, BsmtHalfBath
all_data["BsmtFinSF1"].fillna(0, inplace=True)
all_data["BsmtFinSF2"].fillna(0, inplace=True)
all_data["BsmtFullBath"].fillna(0, inplace=True)
all_data["BsmtHalfBath"].fillna(0, inplace=True)
all_data["MasVnrArea"].fillna(0, inplace=True)
all_data["BsmtUnfSF"].fillna(0, inplace=True)
```

```python
display(all_data[["TotalBsmtSF","BsmtFinSF1", "BsmtFinSF2", "BsmtUnfSF"]][all_data["TotalBsmtSF"].isnull()])
#Since BsmtFinSF1, BsmtFinSF2, BsmtUnfSF=0, TotalBsmtSF should be 0 as well
all_data["TotalBsmtSF"].fillna(0, inplace=True)
```

```python
display (all_data[["GarageYrBlt","GarageType"]][(all_data["GarageYrBlt"].isnull()) & (all_data["GarageType"].isnull()== False)])
# since there is a garage- in column 666 1116 the GarageYrBlt will be the average 
m= all_data["GarageYrBlt"][all_data["GarageType"]=="Detchd"].mean()
all_data["GarageYrBlt"][(all_data["GarageYrBlt"].isnull()) & (all_data["GarageType"].isnull()== False)]=m
```

```python
display (all_data[["GarageCars","GarageArea","GarageType"]][(all_data["GarageCars"].isnull()) & (all_data["GarageType"].isnull()== False)])
# since there is a garage- in column 1116 the GarageCars, GarageArea will be the average
m= all_data["GarageCars"][all_data["GarageType"]=="Detchd"].mean()
all_data["GarageCars"][(all_data["GarageCars"].isnull()) & (all_data["GarageType"].isnull()== False)]=m

m=all_data["GarageArea"][all_data["GarageType"]=="Detchd"].mean()
all_data["GarageArea"][(all_data["GarageCars"].isnull()) & (all_data["GarageType"].isnull()== False)]=m
```

```python
# since there is a garage- in column 1116 the GarageCars, GarageArea will be the average
m= all_data["GarageCars"][all_data["GarageType"]=="Detchd"].mean()
all_data["GarageCars"][(all_data["GarageCars"].isnull()) & (all_data["GarageType"].isnull()== False)]=m

m=all_data["GarageArea"][all_data["GarageType"]=="Detchd"].mean()
all_data["GarageArea"][(all_data["GarageCars"].isnull()) & (all_data["GarageType"].isnull()== False)]=m
```

```python
# else- there is no garage and the value wull be 0
all_data['GarageArea'].fillna(0, inplace=True)
all_data['GarageYrBlt'].fillna(0, inplace=True)
```

```python
display (all_data[["MiscFeature", "MiscVal"]][(all_data["MiscFeature"].isnull() ==False) & (all_data["MiscVal"]==0)])
# we will give the mean value for the specific feature
all_data["MiscVal"][(all_data["MiscFeature"] =="Shead") & (all_data["MiscVal"]==0)]=all_data["MiscVal"][all_data["MiscFeature"]=="Shed"].mean()
all_data["MiscVal"][(all_data["MiscFeature"] =="Othr") & (all_data["MiscVal"]==0)]=all_data["MiscVal"][all_data["MiscFeature"]=="Othr"].mean()
```

```python
display(all_data[["MiscFeature", "MiscVal"]][(all_data["MiscFeature"].isnull()) & (all_data["MiscVal"]!=0)])
for i in all_data["MiscFeature"].unique():
    print  (str( i )+" "+ str(all_data["MiscVal"][all_data["MiscFeature"]==i].mean()))
# since 17000 is the closest to TenC average, we will assume that it should be TenC
all_data["MiscFeature"][(all_data["MiscFeature"].isnull()) & (all_data["MiscVal"]!=0)]="TenC"
```
